# **Matplotlib and Seaborn**

## **1. Library Overview**
######
### **Matplotlib**
* **Description :**
Matplotlib is a core Python library for creating static, animated, and interactive visualizations. Known for its flexibility and detailed customization, it is widely used in data science, machine learning, and scientific research.
* **Unique Features:**
Supports a broad range of plot types, from simple line plots to complex 3D graphs.
Detailed customization of every element within a plot, enabling precise visual control.
Integrates well with libraries like NumPy and Pandas, ideal for data-driven applications.
* **Typical Use Cases:**
Data Analysis: Visualize trends and relationships in data science and research.
Reporting and Documentation: Generate high-quality visuals for publications and presentations.

### **Seaborn**
* **Description:** Seaborn is a high-level statistical data visualization library built on top of Matplotlib. It simplifies the creation of informative and attractive statistical graphics with easy-to-use APIs.
* **Unique Features:**
Automatically handles color palettes and styling to produce visually appealing graphs.
Designed specifically for statistical plots, making it easy to visualize data distributions and relationships.
Enhances readability and aesthetics with its built-in themes and color schemes.
* **Typical Use Cases:**
Exploratory Data Analysis: Quickly generate insights into data distributions and relationships.
Data Science and Machine Learning: Ideal for understanding and interpreting relationships in complex data sets.

## **2. Graph Types and Examples**

Both Matplotlib and Seaborn support a range of graph types suited for different analytical needs. Below are common graph types available in each library, with descriptions, use cases, and example code snippets.

### **Matplotlib Graph Types**
### **1. Line Plot** 
* **Description:** Connects data points with a line, commonly used for time-series data or continuous variables.
* **Use Case:** Ideal for tracking trends over time.
* **Code Example** (using seabornâ€™s flights dataset):


```python
import matplotlib.pyplot as plt
import seaborn as sns

flights = sns.load_dataset("flights")
plt.plot(flights["month"], flights["passengers"], marker='o')
plt.xlabel("Month")
plt.ylabel("Number of Passengers")
plt.title("Monthly Passengers over a Year")
plt.show()
```

## **2.Scatter Plot**

* **Description:** Displays individual data points in a 2D space, used for examining relationships between two variables.
* **Use Case:** Effective for identifying trends, correlations, and outliers.
* **Code Example** (using seabornâ€™s tips dataset):


```python
tips = sns.load_dataset("tips")
plt.scatter(tips["total_bill"], tips["tip"])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total Bill vs Tip Amount")
plt.show()
```

## **3. Bar Chart**

* **Description:** Uses rectangular bars to represent data quantities across categories.
* **Use Case:** Useful for comparing categorical data.
* **Code Example** (using seabornâ€™s titanic dataset):


```python
titanic = sns.load_dataset("titanic")
survived_counts = titanic["class"].value_counts()
plt.bar(survived_counts.index, survived_counts.values)
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.title("Passengers per Class on the Titanic")
plt.show()
```

## **4. Histogram**

* **Description:** Visualizes the distribution of a single variable by grouping data into bins.
* **Use Case:** Suitable for analyzing data distribution and detecting skewness.
* **Code Example** (using seabornâ€™s tips dataset):

```python
plt.hist(tips["total_bill"], bins=20, edgecolor="black")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.title("Distribution of Total Bill Amounts")
plt.show()
```

## **5. Pie Chart**

* **Description:** Represents data as segments of a circle, showing proportions.
* **Use Case:** Useful for visualizing parts of a whole.
* **Code Example** (using seabornâ€™s titanic dataset):


```python
survived_counts = titanic["survived"].value_counts()
plt.pie(survived_counts, labels=["Not Survived", "Survived"], autopct='%1.1f%%', startangle=140)
plt.title("Survival Rate on the Titanic")
plt.show()
```

## **Seaborn Graph Types**

### **1.Line Plot**

* **Description:** Similar to a line plot in Matplotlib but with Seabornâ€™s enhanced styling.
* **Use Case:** Frequently used in time-series data analysis.
* **Code Example** (using flights dataset):


```python
sns.lineplot(data=flights, x="year", y="passengers")
plt.title("Yearly Passengers on Flights")
```

## **2. Scatter Plot**

* **Description:** An enhanced scatter plot with additional styling options for better insights.
* **Use Case:** Examines relationships between two variables.
* **Code Example** (using tips dataset):

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Total Bill vs Tip by Day")
```

## **3.Bar Plot**

* **Description:** Displays bars to compare categories, with Seabornâ€™s styling for readability.
* **Use Case:** Useful for visualizing categorical data distributions.
* **Code Example** (using titanic dataset):


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a bar plot showing average tip per day with gender as hue
sns.barplot(data=tips, x="day", y="tip", hue="sex", estimator=np.mean)

# Set the title
plt.title("Average Tip per Day by Gender")

# Show the plot
plt.show()
```

## **4. Histogram (Distribution Plot)**

* **Description:** Shows data distribution for a single variable, often used with kernel density estimation.
* **Use Case:** Analyzes data spread, skewness, and distribution.
* **Code Example** (using tips dataset):


```python
sns.histplot(data=tips, x="total_bill", kde=True)
plt.title("Distribution of Total Bill with KDE")
```

## **5. Box Plot**

* **Description:** Displays data distribution and variability across categories.
* **Use Case:** Highlights data spread and outliers.
* **Code Example** (using tips dataset):


```python
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Box Plot of Total Bill by Day")
```

| Weaknesses                            | Can be complex and time-consuming to customize                    | Less control over individual visual elements compared to Matplotlib |

Together, these libraries provide a versatile toolkit for data visualization, helping users move seamlessly from data analysis to presentation-ready graphics.
